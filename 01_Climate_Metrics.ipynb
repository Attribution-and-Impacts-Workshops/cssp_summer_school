## 1 — WBGT: Wet Bulb Globe Temperature

### What it measures

WBGT is the standard outdoor heat-stress index. Unlike simple air temperature, it integrates the four environmental factors that together determine how much heat the human body has to shed:

| Component | What it captures |
|-----------|------------------|
| Wet-bulb temperature $T_w$ | Humidity — limits evaporative cooling from sweat |
| Globe temperature $T_g$ | Radiant heat load from sun and surroundings |
| Dry-bulb temperature $T$ | Ambient air temperature |
| (Wind speed) | Implicitly captured through $T_g$ |

### How it is calculated

$$
\text{WBGT} = 0.7 \, T_w + 0.2 \, T_g + 0.1 \, T
$$

The large weight on $T_w$ (70 %) reflects how dominant humidity is in determining heat stress — even a modest rise in humidity at high temperatures can push WBGT into dangerous territory.

In practice, when only gridded model output is available, $T_w$ is approximated from air temperature and relative humidity (e.g. the Stull 2011 formula), and $T_g$ from shortwave radiation and wind speed.

### Threshold guidelines

| WBGT (°C) | Risk level | Typical action |
|-----------|------------|----------------|
| < 28 | Low | Normal activity |
| 28–32 | Moderate | Rest breaks recommended |
| 32–35 | High | Reduce workload / exercise intensity |
| > 35 | Extreme | Stop outdoor activity |

### Impacts studied in attribution

- **Labour productivity** — economic losses from reduced safe working hours for outdoor workers (agriculture, construction)
- **Heat-related morbidity and mortality** — hospital admissions and excess deaths during heatwaves
- **Military and sporting events** — safe limits for training and competition
- **Habitability thresholds** — regions where sustained WBGT > 35 °C would be physiologically unsurvivable without cooling

## 2 — SPEI: Standardised Precipitation-Evapotranspiration Index

### What it measures

SPEI is a **multi-scalar drought index** that quantifies how dry or wet conditions are relative to the long-term average, accounting for both supply (precipitation) and demand (evapotranspiration driven by temperature). It extends the simpler SPI (which uses only precipitation) by including the effect of warming on water loss from soils and plants.

A key advantage of SPEI over SPI in a changing climate is that as temperatures rise, evaporative demand increases even without any change in precipitation — meaning droughts can intensify purely through a warming signal.

### How it is calculated

**Step 1 — Climatic water balance:**

$$
D_i = P_i - \text{PET}_i
$$

where $P_i$ is monthly precipitation and $\text{PET}_i$ is potential evapotranspiration. PET is commonly estimated using the **Penman–Monteith** (temperature, radiation, wind speed, and relative humidity) method.

**Step 2 — Accumulate at a chosen time scale $k$:**

$$
D_i^k = \sum_{j=0}^{k-1} D_{i-j}
$$

Time scales of 1, 3, 6, 12, and 24 months capture different aspects of drought.

**Step 3 — Fit a probability distribution and standardise:**

The accumulated $D^k$ series is fitted to a **log-logistic distribution** and transformed to a standard normal score. The result is SPEI, with:

| SPEI value | Category |
|------------|----------|
| > +2 | Extremely wet |
| +1 to +2 | Moderately wet |
| −1 to +1 | Near normal |
| −1 to −2 | Moderate drought |
| < −2 | Extreme drought |

### Impacts studied in attribution

- **Agricultural drought** — crop failures, reduced yields, food security
- **Hydrological drought** — low river flows, reservoir depletion, hydropower shortfalls
- **Ecosystem stress** — vegetation die-off, tree mortality, peatland drying
- **Wildfire** — dry vegetation as fuel (SPEI often used alongside FWI)
- **Compound events** — drought + heat (high WBGT + negative SPEI) amplify both agricultural and health impacts

## 3 — CDD: Cooling Degree Days

### What it measures

Cooling Degree Days (CDD) measure the **cumulative excess of daily temperature above a baseline** — a threshold below which buildings are assumed not to need mechanical cooling. CDD is a proxy for **air-conditioning energy demand**: the higher the CDD, the more cooling is needed.

CDD is closely linked to electricity grid stress, especially during summer heatwaves when demand can spike simultaneously across a region.

### How it is calculated

For each day $i$:

$$
\text{CDD}_i = \max\bigl(0,\; \bar{T}_i - T_{\text{base}}\bigr)
$$

where $\bar{T}_i$ is the mean daily temperature and $T_{\text{base}}$ is the cooling baseline.

Seasonal or annual CDD is the sum over all days in the period:

$$
\text{CDD}_{\text{season}} = \sum_{i} \max\bigl(0,\; \bar{T}_i - T_{\text{base}}\bigr)
$$

**Common baseline temperatures:**

| Baseline | Context |
|----------|---------|
| 18 °C | European standard (IEA, Eurostat) |
| 22 °C | Tropical / subtropical contexts |
| 24 °C | Some national energy standards |

The choice of baseline matters: a higher baseline makes CDD more sensitive to extreme heat rather than moderate warmth.

### Impacts studied in attribution

- **Electricity demand and grid stress** — peak demand exceeding grid capacity, brownouts and blackouts
- **Energy poverty** — households that cannot afford to run air conditioning during extreme heat
- **Urban heat island** — cities with more concrete and less vegetation accumulate higher CDD
- **Health** — indoor temperature determines heat-related illness for those without cooling access
- **Infrastructure costs** — increased capital expenditure on cooling systems as baseline climate warms

## 4 — RX1day: Maximum 1-Day Precipitation

### What it measures

RX1day is the **annual (or seasonal) maximum of daily precipitation** — the single wettest day in a given year. It is one of the core indices defined by the ETCCDI (Expert Team on Climate Change Detection and Indices) and is widely used as a proxy for **heavy rainfall events** that cause flash flooding and infrastructure damage.

Because climate change is expected to intensify the water cycle (the Clausius–Clapeyron relation: ~7 % more moisture per degree of warming), RX1day is a particularly clear fingerprint of human influence on the hydrological cycle.

### How it is calculated

$$
\text{RX1day} = \max_{i \in \text{year}} \bigl( P_i \bigr)
$$

where $P_i$ is the total precipitation on day $i$. The calculation is straightforward:
1. Take all daily precipitation values for the year (or season)
2. Return the maximum

For attribution, the annual series of RX1day values is typically fitted to a **GEV (Generalised Extreme Value)** or **Gumbel** distribution, as these are the theoretically correct distributions for block maxima.

**Related indices:**

| Index | Definition |
|-------|------------|
| RX1day | Max 1-day precipitation |
| RX5day | Max consecutive 5-day precipitation |
| R95p | Total precipitation above 95th percentile |
| PRCPTOT | Total annual precipitation |

RX5day is more relevant for river flooding (catchment accumulation), while RX1day is more relevant for flash flooding and urban drainage.

### Impacts studied in attribution

- **Flash flooding** — urban areas overwhelmed by drainage systems designed for a past climate
- **Landslides** — intense rainfall on saturated or deforested slopes
- **River flooding** — combined with antecedent soil moisture from SPEI
- **Crop damage** — physical damage to plants and soil erosion
- **Infrastructure** — road washouts, bridge failures, sewer overflow

> **Attribution context:** RX1day events in a warmer world are expected to become more intense (higher magnitude for the same return period) even if their frequency does not change, because a warmer atmosphere holds more moisture available for precipitation.

## 5 — FWI: Fire Weather Index

### What it measures

The **Canadian Forest Fire Weather Index (FWI) System** is the most widely used operational fire danger rating system globally. The FWI integrates meteorological conditions to estimate the **danger of a fire starting, spreading, and being difficult to control**. It is a purely weather-based index — it does not directly account for vegetation type or ignition source.

### How it is calculated

The FWI System is built from a hierarchy of six sub-indices, each capturing a different aspect of fire danger:

```
Inputs: Temperature, Relative Humidity, Wind Speed, 24h Precipitation
          │              │                              │
          ▼              ▼                              ▼
       FFMC            DMC                            DC
  (surface litter)  (shallow organic)          (deep organic)
          │              │                              │
          ▼              └──────────────────────────────┘
         ISI                         BUI
  (rate of spread)           (total fuel available)
          │                          │
          └──────────────────────────┘
                       ▼
                      FWI
                (overall fire danger)
```

| Sub-index | Full name | Driven by |
|-----------|-----------|----------|
| FFMC | Fine Fuel Moisture Code | Temperature, RH, wind, rain |
| DMC | Duff Moisture Code | Temperature, RH, rain |
| DC | Drought Code | Temperature, rain (long memory) |
| ISI | Initial Spread Index | FFMC × wind speed |
| BUI | Buildup Index | DMC + DC |
| **FWI** | **Fire Weather Index** | **ISI × BUI** |

The DC has a very long memory (months), making it sensitive to antecedent drought — this is why FWI overlaps with the drought signal captured by SPEI.

**FWI danger classes:**

| FWI | Class |
|-----|-------|
| 0–5 | Very Low |
| 5–12 | Low |
| 12–20 | Moderate |
| 20–30 | High |
| 30–50 | Very High |
| > 50 | Extreme |

These classes are approximate — thresholds are calibrated regionally.

### Impacts studied in attribution

- **Area burned** — FWI is correlated with the total area that burns in a fire season
- **Air quality** — smoke from wildfires causes PM2.5 exceedances far downwind
- **Carbon emissions** — forest and peatland fires release stored carbon
- **Biodiversity loss** — habitat destruction, especially in fire-sensitive ecosystems
- **Infrastructure and housing** — wildland–urban interface fires
- **Compound risk** — high FWI combined with drought (negative SPEI) and heat (high WBGT) creates conditions for the most dangerous and fast-moving fires

> **Attribution context:** Attribution studies typically ask whether the **number of high-FWI days** or the **seasonal mean FWI** was made more likely by climate change. Because FWI responds to temperature, humidity, and antecedent precipitation, its attribution signal combines the heat and drought fingerprints.

## Summary: Metrics at a glance

| Metric | Primary driver | Timescale | Key impact sector | Distribution used in attribution |
|--------|---------------|-----------|-------------------|----------------------------------|
| WBGT | Temperature + humidity + radiation | Daily | Health, labour | Normal / GEV |
| SPEI | Precipitation − PET | Monthly (multi-scalar) | Agriculture, water, wildfire | Normal (by construction) |
| CDD | Temperature | Seasonal cumulative | Energy demand | Normal / GEV |
| RX1day | Precipitation | Annual maximum | Flooding, infrastructure | GEV / Gumbel |
| FWI | Temperature + humidity + wind + rain | Daily / seasonal | Wildfire | GEV / Gamma |

### Compound events

These metrics do not operate in isolation. Some of the most severe impacts arise from **compound events** — when two or more hazards occur simultaneously or in close succession:

- **Hot and dry** (high WBGT + negative SPEI): amplified heat stress and fire risk
- **Drought then heavy rain** (negative SPEI preceding high RX1day): landslides on dry soils
- **Heat and humidity** (high WBGT during high CDD period): health and energy demand peaks together

When you run attribution on any one of these metrics, keep in mind the meteorological context — a single-metric PR can underestimate the true societal risk if the event was compound.
