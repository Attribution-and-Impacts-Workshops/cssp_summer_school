# Setting Up Your `climattr` Environment on JASMIN

This notebook walks through creating a Python virtual environment on the **JASMIN Notebook Service** and installing the `climattr` package from GitHub.

**Run this notebook once** before opening `WBGT_Attribution.ipynb`.

---

### Why a virtual environment?

JASMIN's shared Python environments already have some packages installed that could have a conflict with `climattr` package. A virtual environment gives you an isolated space to install your own packages.

---

### How to run the setup steps

The commands below need to be run in a **terminal**, not as notebook cells, because `conda activate` does not persist across notebook cell executions.

Open a terminal in JupyterLab: **File → New → Terminal**, then follow the steps.

---
## Step 1 — Open JASMIN Notebooks and clone the repository

1. Go to [https://notebooks.jasmin.ac.uk](https://notebooks.jasmin.ac.uk) and log in
2. Open a new terminal: **File → New → Terminal**
3. Clone the course repository:

```bash
git clone https://github.com/Attribution-and-Impacts-Workshops/cssp_summer_school.git
```

---
## Step 2 — Create the virtual environment

```bash
mkdir -p ~/nb_envs
python -m venv ~/nb_envs/climattr-env
```

---
## Step 3 — Activate the environment and install `climattr`

```bash
source ~/nb_envs/climattr-env/bin/activate
pip install ipykernel
pip install git+https://github.com/rafaelcabreu/climattr@master
```

---

---
## Step 4 — Register the environment as a Jupyter kernel

Still in the same terminal with `climattr-env` activated:

```bash
python -m ipykernel install --user --name=climattr-env
```

You only need to do this once. The kernel will persist across JASMIN sessions.

---
## Step 5 — Select the kernel in JupyterLab

After registering, switch this notebook (and any other notebook) to use the new kernel:

1. Click the kernel name in the top-right corner of the notebook (e.g. *"Python 3 (ipykernel)"*)
2. Select **"climattr-env"** from the list
3. Click **Select**

> If **"climattr-env"** does not appear, refresh the browser page and try again.

---
## Step 6 — Verify the installation

**Switch to the "climattr-env" kernel** (Step 5 above), then run the cell below.

````python
import climattr as eea
import xarray as xr
import numpy as np
import scipy.stats as stats
import matplotlib
import cartopy

print(f"climattr  : OK")
print(f"xarray    : {xr.__version__}")
print(f"numpy     : {np.__version__}")
print(f"matplotlib: {matplotlib.__version__}")
print(f"cartopy   : {cartopy.__version__}")
print()
print("All imports successful — you are ready to open WBGT_Attribution.ipynb")
````
